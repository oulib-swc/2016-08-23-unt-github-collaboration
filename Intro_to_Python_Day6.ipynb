Write a function to multiply two given 2D-arrays of same size element-by-element if they have the same shape.

```python
def safe_multiply(array1, array2):
    if array1.shape == array2.shape:
        return array1 * array2
    else:
        print("these two array have different shape, i can't multiply element-wise here")
        return None
```

```python
array1 = np.array([1, 2, 3, 4])
array2 = np.linspace(1, 20, 5)
# print(array1 * array2)
print(safe_multiply(array1, array2))
```

### Numpy Array Slicing

In Numpy the slicing for 1-D array is almost the same as slicing lists. 

```python
#slicing arrays 
#[start:end:step]
#including the start index but excluding the end index 
sample_np = np.array([1, 3, 5, 7, 9, 11])
print(sample_np[1:3])
print(sample_np.shape) 
```

```python
#omitting some parameters in an array 
#omit the start - start from index 0 
#omit the step - step by 1 
#omit the end - until the end 
print(sample_np[1:])
type(sample_np[1:])
```

```python
print(sample_np[:3])
```

```python
#negative indexing 
print(sample_np[-3:-1])
```

```python
#using step 
print(sample_np[:3:2])
print(sample_np[: :2])
```

We can also slice 2-D array by specifying the dimension. 

```python
#slicing 2-D array 
sample_2d = np.array([[1,3,5,7,9],[2,4,6,8,10]])
print(sample_2d[1,1:4:2])
```

```python
#get elements from both dimensions 
sliced = sample_2d[0:2,3]
#[:,3]
print(sliced)
sliced.dtype
#the sliced part becomes a new array
sliced.shape
```

```python
#the outcome is a 2D-array
print(sample_2d[0:2, 1:4])
```

### Making a copy of the array 

```python
import numpy as np
#copying the array use .copy()
#mutation of original array does not affect the copied array 
original = np.array([0, 10, 30, 5])
copy = original.copy()
original[2] = -5

print(original, copy)
```

```python
#Numpy has another command called view 
#What would happen if we use view? 
#the viewed array will be affected by the change of original arrays
new = original.view()
original[1] = 2
print(original, new)
```

```python
a = [1,2,3]
# b = a # not creating a new array
b = [x for x in a] # creating a new array
b[2] = 5
print("a: ", a, "b: ", b)
```

##### The slice of an array is a view, which means that modifying subarrays will affect the original array.

```python
#very important
ori = np.array([[1,3,5,6],[2,4,6,7]])
sub_array = ori[:2, 1:3]
print(sub_array)
sub_array[1,1] = 100 
#the original array is also modified. 
print(sub_array)
print(ori)
```

### Joining arrays
np.concatenate() and np.stack()

```python
#same as strings and lists, we can concatenate arrays using np.concatenate
arr1 = np.array([1,2])
arr2 = np.array([3,4])

new_arr = np.concatenate((arr1,arr2))
print(new_arr)
```

```python
#2D-array 
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])
#join by rows
#in Python axis = 1 means rows, axis = 0 means columns
new_arr = np.concatenate((arr1,arr2), axis = 1)
print(new_arr)
```

```python
#stack() is similar to concatenate(), which is used for joining arrays:
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
#default value of axis is 0
new_arr = np.stack((arr1, arr2), axis = 0)
print(new_arr)
```

```python
#two other methods of stacking: stacking along rows vs columns 
new_arr_h = np.hstack((arr1, arr2))
print(new_arr_h)
new_arr_v = np.vstack((arr1, arr2))
print(new_arr_v)
```

## Pandas library

```python
import pandas as pd 
from pandas import DataFrame, Series
#Series is used to deal with 1-D data and DataFrames are usded to handle 2-D data.
```

#### Handling missing values

```python
#Pandas handling missing value
#for A,B,C,D, we did not assign values to these four indexes. so Pandas automatically assign them to NAN
#which represents "Not a Number". 
#Let's try a different case: 
student = {'George':'Oakland', 'Lisa':'Washtenaw', 'Bill':'Saint Clair', 'Kim':'Macomb'}
obj2 = pd.Series(student)
obj2
obj2.index
#panda set a list of string data as object
```

```python
new_student = ['Anne', 'Steve']
obj3 = pd.Series(student, index=new_student)
obj3
```

```python
#used to be int and None, while Pandas tranfroms it into floating point numbers and NAN
#in this case, if your data contain missing data, Pandas will convert all the data into floating point numbers
#NAN is not equivalent to None
numbers = [1, 2, None]
pd.Series(numbers)
```

```python
#NAN is not equivalent to None!
np.nan == None
```

```python
#use np.isnan() to assess whether there is a missing data
np.isnan(np.nan)
#NAN is treated as a numeric value
#can we use np.nan == np.nan?
np.nan == np.nan
```

```python
#omit some items in the dictionary while creating series object 
#watermelon is omitted because this key is not included in index, but we add a new one, cherry.
#Cherry does not have a value so it is handled using "NAN"
fruit ={'apple':2.5, 'banana':.99, 'watermelon':2.3}
obj4 = pd.Series(fruit, index = ['apple','banana','cherry'])
obj4
```

### Querying series object 

```python
#loc[''], iloc[index number]
#create a dictionary
#square brakets because loc or iloc are attributes
#not parentheses
obj = {'Sam': 'Ohio', 'Peter':'Michigan', 'Kim': 'New York','Stephen':'Atlanta'}
name = pd.Series(obj)
name.iloc[2] #zero-indexed
name.loc['Sam'] #get the value corresponding with the index
```

```python
#index operator, you can also directly use square brackets to get the values
print(name[2],name['Sam'])
```

```python
#while, when you have integers as indexes, it can be confusing for python to understand 
#whether you are using indexing number or "keys"
#Therefore, using .iloc[] or .loc[] is more recommended.
obj1 = {1:'Alex', 2:'Goerge', 99:'Linda', 3:'Peter'}
name = pd.Series(obj1)
#the following statements can create confusion
name.iloc[2]
name[2]
```

#### Broadcasting

```python
#The classic way of increasing values by 2 for all the elements 
data = pd.Series(6, index = range(4))
for label, value in data.iteritems():
    data.loc[label] = value +2
data.head()
```

##### We can have indexes of same values in Pandas.

```python
#As we showed above, the values could be the same across different indexes (the example of scalar)
#the indexes could also be the same 
#Recall what would happen if we have same keys in dictionary...
grade = pd.Series([80, 90, 100], index =['Alex','Alex','Alex'])
grade
```

```python
#A.append(B) - Append B on A Series object
#unlike list, append() function in Pandas does not modify original list in place, instead, we need to create
#a new Series object and combine the two Series object together.
class_grade = pd.Series([90, 95, 87, 80], index = ['George', 'Susan', 'Lisa', 'Kim'])
all_class_grade = class_grade.append(grade)
all_class_grade
```

```python
#What if we try to append two Series objects with different data types? 
#Pandas will figure out a way to combine these two series objects together and make the data type
#to be the more homogeneous one
course = pd.Series(['math', 'psychology','music'], index = ['Alex','Alex','Alex'])
all_students = class_grade.append(course)
all_students
```
